# Производная и с чем её едят

## 1 Производная

Пусть $f(x)$ — функция одной переменной. Её производная в точке $x_0$ определяется как предел отношения приращений при стремлении приращения аргумента к нулю:

$$
f'(x_0)=\lim_{\Delta x\to 0}\frac{f(x_0+\Delta x)-f(x_0)}{\Delta x}.
$$

Этот предел, если он существует, задаёт скорость изменения $f$ в точке.  

### 1.1 Правила (как её вычислять)

Из определения следуют стандартные правила (получаемые пределами):

- $(u+v)' = u' + v'$ — производная суммы  
- $(u \cdot v)' = u'v + uv'$ — правило произведения  
- $(u(v(x)))' = u'(v(x)) \cdot v'(x)$ — цепное правило  


### 1.2 Примеры

Для $f(x)=\sin(x)$:

$$
f'(x)=\cos(x)
$$

Для $f(x)=e^x$:

$$
f'(x)=e^x
$$

Эти формулы можно проверить из определения через предел.

### 1.3 Геометрический смысл производной

**Идейно:** производная моментальная скорость изменения функции. 

Производная функции в точке показывает наклон касательной к графику функции в этой точке (уравнение прямой: $y = kx + b$).

Иначе говоря, это скорость изменения функции: насколько быстро изменяется значение функции при малом изменении аргумента.

Если производная положительна, функция возрастает; если отрицательна — убывает; если равна нулю — в точке возможен экстремум.

```python
%conda install numpy
%conda install matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**2)

# Define x values
x = np.linspace(-7, 7, 400)

# Define function and its derivative
y = [f(_) for _ in x]

x_0 = -5 
y_0 = f(x_0)

k_1 = (f(x_0 + 5) - f(x_0)) / (5)
k_2 = (f(x_0 + 3) - f(x_0)) / (3)
k_3 = (f(x_0 + 0.5) - f(x_0)) / (0.5)
k_4 = (f(x_0 + 0.01) - f(x_0)) / (0.01)

l_1 = y_0 + k_1*(x[:200] - x_0)
l_2 = y_0 + k_2*(x[:200] - x_0)
l_3 = y_0 + k_3*(x[:200] - x_0)
l_4 = y_0 + k_4*(x[:200] - x_0)


# Plot the function
plt.figure()
plt.plot(x, y)
plt.plot(x[:200], l_1)
plt.plot(x[:200], l_2)
plt.plot(x[:200], l_3)
plt.plot(x[:200], l_4)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

## 2. Численные методы вычисления производных

Когда аналитическую производную функции вычислить сложно или функция задана «чёрным ящиком» (например, только через программу), используются численные методы.

В основе простейших методов лежит представление производной как предел отношения приращений. Приблизительно можно использовать разностные формулы с малым конечным шагом $h$.


### 2.1 Методы конечных разностей

#### Прямая (вперёд) разность

$$
f'(x) \approx \frac{f(x+h) - f(x)}{h}
$$

Такая оценка даёт погрешность порядка $O(h)$: она пропорциональна $h$.

При этом:
- ошибка уменьшается при уменьшении $h$,
- но возникают ошибки округления.


#### Обратная (назад) разность

$$
f'(x) \approx \frac{f(x) - f(x - h)}{h}
$$

Также имеет точность порядка $O(h)$ и обычно используется симметрично с прямой разностью.

#### Центральная разность

$$
f'(x) \approx \frac{f(x+h) - f(x-h)}{2h}
$$

Эта формула более точная: её погрешность порядка $O(h^2)$.

Центральная разность даёт квадратичное уменьшение ошибки по сравнению с прямой, то есть лучше приближает $f'(x)$ при том же $h$.

### 2.2 Пример

Для $f(x)=\sin(x)$ в точке $x=1$ можно численно вычислить производную и сравнить с точным значением:

$$
f'(1) = \cos(1)
$$

Приближения:

- Прямая разность (вперёд):
$$
\frac{\sin(1+h)-\sin(1)}{h}
$$

- Центральная разность:
$$
\frac{\sin(1+h)-\sin(1-h)}{2h}
$$

```python
import numpy as np

def forward_diff_sin(x, h):
    return (np.sin(x+h) - np.sin(x)) / h

def central_diff_sin(x, h):
    return (np.sin(x+h) - np.sin(x-h)) / (2*h)

x0 = 1.0
h = 1e-3
print("Forward:", forward_diff_sin( x0, h), "True:", np.cos(x0))
print("Central:", central_diff_sin(x0, h), "True:", np.cos(x0))
```

### 2.3 Выбор шага $h$ и ошибки

При выборе шага $h$ возникает компромисс между различными типами ошибок:

- При слишком большом $h$ увеличивается погрешность аппроксимации.

- При слишком малом $h$ начинают доминировать ошибки округления  
  и «неустранимые» погрешности вычислений (машинная точность),  
  из-за чего абсолютная ошибка может расти.


Таким образом, задача численного дифференцирования является некорректно поставленной:

- погрешность аппроксимации стремится к нулю при $h \to 0$,
- но вычислительные ошибки (связанные с потерей точности) растут.

Поэтому на практике выбирают **оптимальный промежуточный шаг** $h$.

Обычно:
- анализируют поведение ошибки,
- подбирают $h$ адаптивно.

## 3 Градиентный спуск

### 3.1 Одномерный случай

**Интуиция:** стоим на холме в тумане и хочем спуститься вниз. 
Мы не видим всю поверхность, но чувствуешь наклон, а значит идём вниз по самому крутому спуску.
Это и есть **градиентный спуск**.


Градиент — это направление наибольшего роста функции. В одномерном случае просто идём по производной.
Чтобы найти минимум, идём в противоположную сторону.

Формула обновления:

$$
x_{n+1} = x_{n} - lr \cdot \left. \frac{d f}{ d x} \right|_{x=x_n} = x_{n} - lr \cdot f'(x_{n}) 
$$

где:
- lr — learning rate (шаг)
- $x_n$ — значение переменной
- $f'(x_n)$ — значение производной

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def grad(f,x,dx):
    return (f(x+dx) - f(x))/(dx)

def gradient_descent(f, x, lr, dx):
    x_new = x - lr * grad(f, x, dx)
    return x_new

x = np.linspace(-7, 7, 400)

y = [f(_) for _ in x]

x_0 = -5 
y_0 = f(x_0)

route_x = [x_0]
route_y = [y_0]

for i in range(30):
    x_new = gradient_descent(f, x_0, 0.1, 0.1)
    route_x.append(x_new)
    route_y.append(f(x_new))
    x_0 = x_new


# Plot the function
plt.figure()
plt.plot(x, y)
plt.plot(route_x, route_y, '-o', color='red', linewidth=1, markersize=8)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

def f(x):
    return x**2

def grad(f, x, dx):
    return (f(x + dx) - f(x)) / dx

def gradient_descent(f, x, lr, dx):
    return x - lr * grad(f, x, dx)

x = np.linspace(-7, 7, 400)
y = [f(_) for _ in x]

x_0 = -5

route_x = [x_0]
route_y = [f(x_0)]

for i in range(30):
    x_new = gradient_descent(f, x_0, 0.1, 0.1)
    route_x.append(x_new)
    route_y.append(f(x_new))
    x_0 = x_new

fig, ax = plt.subplots()

ax.plot(x, y)
point, = ax.plot([], [], 'o')
path, = ax.plot([], [], '-')

ax.set_title("Gradient Descent Animation")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)

def update(frame):
    xs = route_x[:frame+1]
    ys = route_y[:frame+1]
    
    point.set_data(xs[-1:], ys[-1:])
    path.set_data(xs, ys)
    
    return point, path

anim = FuncAnimation(fig, update, frames=len(route_x), interval=300)

HTML(anim.to_jshtml())
```

### 3.2 Переход от одномерного случая к двумерному

В одномерном случае мы рассматривали функцию $y = f(x)$ и использовали обычную производную, чтобы понять, куда двигаться.

- если производная положительная → идём влево  
- если отрицательная → идём вправо  

Теперь рассмотрим функцию двух переменных $z = f(x, y)$.

В этом случае одного числа уже недостаточно, чтобы описать наклон.  
У функции есть наклон сразу в двух направлениях:
- вдоль оси $x$
- вдоль оси $y$

Поэтому вместо одной производной появляется **градиент** — вектор из частных производных:

$$
\nabla \left. f(x, y) \right|_{x=x_0, y=y_0} = \left( \left. \frac{\partial f}{\partial x} \right|_{x=x_0, y=y_0}, \left. \frac{\partial f}{\partial y} \right|_{x=x_0, y=y_0} \right)
$$

Градиент показывает направление **наибольшего роста функции**.

Чтобы найти минимум, нужно двигаться в противоположную сторону — против градиента.

Правило обновления теперь выглядит так:

$$
x_{n+1} = x_{n} - lr \cdot \left. \frac{\partial f}{\partial x} \right|_{x=x_n, y=y_n}
$$

$$
y_{n+1} = y_{n} - lr \cdot \left. \frac{\partial f}{\partial y} \right|_{x=x_n, y=y_n}
$$

Интуиции: 

- В 1D: идём по линии  
- В 2D: идём по поверхности  

--- 

**Вопрос:** как считать производную? 

**Ответ:** да ровно также:

$$ \left. \frac{\partial f}{\partial x} \right|_{x=x_n, y=y_n} = \left. \frac{f(x_n + h, y_n) - f(x_n, y_n)}{ h} \right. $$

$$ \left. \frac{\partial f}{\partial y} \right|_{x=x_n, y=y_n} = \left. \frac{f(x_n, y_n + h) - f(x_n, y_n)}{ h} \right. $$

```python
import numpy as np
import matplotlib.pyplot as plt

# функция
def f(x, y):
    return x**2 + y**2

# численный градиент
def grad(f, x, y, dx):
    dfdx = (f(x + dx, y) - f(x, y)) / dx
    dfdy = (f(x, y + dx) - f(x, y)) / dx
    return np.array([dfdx, dfdy])

# шаг градиентного спуска
def gradient_descent(f, x, y, lr, dx):
    g = grad(f, x, y, dx)
    x_new = x - lr * g[0]
    y_new = y - lr * g[1]
    return x_new, y_new

# сетка для поверхности
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# начальная точка
x0, y0 = -5, 4

route_x = [x0]
route_y = [y0]
route_z = [f(x0, y0)]

# градиентный спуск
for i in range(30):
    x_new, y_new = gradient_descent(f, x0, y0, lr=0.1, dx=0.01)
    
    route_x.append(x_new)
    route_y.append(y_new)
    route_z.append(f(x_new, y_new))
    
    x0, y0 = x_new, y_new

# --- ВИЗУАЛИЗАЦИЯ ---

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# поверхность
ax.plot_surface(X, Y, Z, alpha=0.6)

# путь спуска
ax.plot(route_x, route_y, route_z, '-o')

ax.set_title("Gradient Descent on f(x,y)=x^2+y^2")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
from IPython.display import HTML

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# функция
def f(x, y):
    return x**2 + y**2

# численный градиент
def grad(f, x, y, dx):
    dfdx = (f(x + dx, y) - f(x, y)) / dx
    dfdy = (f(x, y + dx) - f(x, y)) / dx
    return np.array([dfdx, dfdy])

# шаг градиентного спуска
def gradient_descent(f, x, y, lr, dx):
    g = grad(f, x, y, dx)
    return x - lr * g[0], y - lr * g[1]

# поверхность
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# начальная точка
x0, y0 = -5, 4

route = [(x0, y0, f(x0, y0))]

# считаем путь заранее
for _ in range(40):
    x0, y0 = gradient_descent(f, x0, y0, lr=0.7, dx=0.01)
    route.append((x0, y0, f(x0, y0)))

# --- ВИЗУАЛИЗАЦИЯ ---
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.6)

point, = ax.plot([], [], [], 'o')
path, = ax.plot([], [], [], '-')

# функция обновления
def update(frame):
    xs = [p[0] for p in route[:frame+1]]
    ys = [p[1] for p in route[:frame+1]]
    zs = [p[2] for p in route[:frame+1]]
    
    point.set_data(xs[-1:], ys[-1:])
    point.set_3d_properties(zs[-1:])
    
    path.set_data(xs, ys)
    path.set_3d_properties(zs)
    
    return point, path

anim = FuncAnimation(fig, update, frames=len(route), interval=200)

ax.set_title("Gradient Descent Animation")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

HTML(anim.to_jshtml())
```

### 3.3 Критерий остановки градиентного спуска

Градиентный спуск является итерационным алгоритмом, поэтому необходимо определить условие, при котором процесс следует остановить.

На практике используются несколько основных критериев остановки:

1. Ограничение по числу итераций  
   Алгоритм выполняется заданное количество шагов. Это самый простой способ, однако он не гарантирует достижение минимума.

2. Малое изменение аргумента  
   Если разница между значениями на соседних итерациях становится меньше заданного порога,
   $$
   \|x_{k+1} - x_k\| < \varepsilon,
   $$
   считается, что алгоритм сошелся.

3. Малое значение градиента  
   Если норма градиента становится достаточно маленькой,
   $$
   \|\nabla f(x_k)\| < \varepsilon,
   $$
   это означает, что точка находится близко к минимуму.

4. Малое изменение значения функции  
   Если изменение функции между итерациями мало,
   $$
   |f(x_{k+1}) - f(x_k)| < \varepsilon,
   $$
   дальнейшие улучшения незначительны.

На практике часто используют комбинацию нескольких критериев, например ограничение на число итераций и условие на норму градиента.

###  3.4 Зачем нужны производные? 

Градиентный спуск — это один из основных алгоритмов оптимизации, который широко используется в машинном обучении.

##### Зачем он нужен?

Во многих задачах машинного обучения требуется **настроить параметры модели**, чтобы она как можно лучше описывала данные.

Для этого вводится функция ошибки (или функция потерь), которая показывает, **насколько плохо модель работает**.

Задача обучения сводится к следующему:

> Найти такие параметры модели, при которых функция ошибки минимальна.

##### Пример: линейная регрессия

Пусть модель имеет вид:

$$
y = wx + b
$$

Наша цель — подобрать параметры $w$ и $b$, чтобы предсказания модели были как можно ближе к реальным данным.

Для этого используется функция потерь, например среднеквадратичная ошибка:

$$
L(w, b) = \frac{1}{n} \sum (y_{pred} - y_{true})^2
$$

##### Как применяется градиентный спуск

1. Инициализируем параметры $w$ и $b$ (например, случайными значениями)

2. Считаем градиент функции потерь по параметрам:
   - $\frac{\partial L}{\partial w}$
   - $\frac{\partial L}{\partial b}$

3. Обновляем параметры:

$$
w = w - lr \cdot \frac{\partial L}{\partial w}
$$

$$
b = b - lr \cdot \frac{\partial L}{\partial b}
$$

4. Повторяем процесс много раз


**Интуиция:**
- функция потерь — это "поверхность ошибок"
- параметры модели — это точка на этой поверхности
- градиент показывает, как быстрее всего увеличить ошибку
- значит, чтобы уменьшить ошибку, нужно двигаться в противоположную сторону


**Важные моменты:**
- слишком большой шаг обучения → алгоритм может не сойтись  
- слишком маленький → обучение будет очень медленным  
- возможны локальные минимумы  

**Вывод:**
Градиентный спуск — это универсальный инструмент, который позволяет обучать модели, минимизируя функцию ошибки и улучшая качество предсказаний.
