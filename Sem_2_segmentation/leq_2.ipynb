### Стемминг и лематизация

Следующий этап обработки текста — приведение слов к нормализованной форме. Эта задача сложнее простого разделения на токены: регулярных выражений здесь недостаточно, требуются либо морфологические словари, либо более продвинутые методы, включая нейронные сети.

Однако ещё в прошлом веке для английского языка был предложен упрощённый способ унификации словоформ — стемминг. Это, по сути, исторический подход: в 1980-х годах, когда компьютеры не имели графических интерфейсов, а автоматический морфологический анализ был практически недоступен, Мартин Портер разработал алгоритм, отсекающий окончания и суффиксы от псевдоосновы слова. Этот алгоритм, известный как «стеммер Портера», до сих пор доступен для ряда европейских языков, включая русский (в более новой версии Snowball).

Как любой правило-ориентированный метод, стеммер Портера работает неидеально и допускает ошибки. В библиотеке NLTK представлено несколько реализаций стеммеров, например:

1. PorterStemmer
2. SnowballStemmer
3. LancasterStemmer
4. RegexpStemmer
5. RSLPStemmer

Мы рассмотрим PorterStemmer и SnowballStemmer.

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer

ps = PorterStemmer()
ss = SnowballStemmer('english')

plurals = ['caresses', 'flies', 'dies', 'mules', 'denied',
            'died', 'agreed', 'owned', 'humbled', 'sized',
            'meeting', 'stating', 'siezing', 'itemization',
            'sensational', 'traditional', 'reference', 'colonizer',
            'plotted']

for x in plurals: 
    print(f"Porter: {ps.stem(x):<15} Snowball: {ss.stem(x):<15}")
```

Более точной и лингвистически корректной альтернативой стеммингу является лемматизация. В отличие от алгоритмического "отрубания" окончаний, цель лемматизатора — привести слово к его нормальной словарной форме, или лемме. Например, для глаголов это инфинитив ("делал" → "делать"), для существительных — именительный падеж единственного числа ("столами" → "стол").

В библиотеке NLTK эту задачу решает WordNetLemmatizer. Его работа основана не на правилах, а на обращении к обширной лингвистической базе данных WordNet, которая хранит связи между словами, включая их основные формы. Для корректной работы этому лемматизатору часто необходимо знать часть речи слова (например, глагол это или существительное), иначе результат может быть неточным.

Ключевое ограничение встроенного WordNetLemmatizer — его языковая специфичность. Поскольку он завязан на словарь WordNet, созданный для английского языка, он не подходит для обработки русского или других языков.

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

plurals = ['increases', 'playing', 'cats']

for x in plurals: 
    print(f"Porter: {ps.stem(x):<15} Snowball: {ss.stem(x):<15} WordNetLemmatizer (pos=v): {wnl.lemmatize(x, pos='v'):<15}  WordNetLemmatizer (pos=v): {wnl.lemmatize(x, pos='a'):<15}")
```

В NLTK есть списки таких слов для разных языков:

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

example_sent = 'При обработке текстов для решения задач NLP, особенно если используются классические алгоритмы машинного обучения (современные нейронки извлекают признаки из всего, им часто лучше, чтобы текст сохранялся в исходном виде), бывает нужно выкинуть слишком распространенные и малозначимые слова: союзы, предлоги и т.п. В NLTK есть списки таких слов для разных языков'
stop_words = set(stopwords.words('russian'))
print('Стоп-слова:', stop_words)
word_tokens = word_tokenize(example_sent)
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
print(filtered_sentence)
```

### Алгоритм MaxMatch (Максимальное соответствие) 

Это простой алгоритм для сегментации текста (токенизации), например, для разделения слитного предложения на слова в языках без явных разделителей (пробелов) (китайский и аналогичные языки, где нет пунктуации) или для обработки текста в NLP.

Основная идея: Начиная с начала строки, всегда выбирать самое длинное слово из словаря, которое соответствует началу оставшейся части строки.

Пошаговый процесс:

1. Начинаем с начала строки. Указатель стоит на первом символе.
2. Ищем самое длинное слово. Среди всех слов в доступном словаре находим самое длинное, которое совпадает с символами строки, начиная с текущей позиции.
3. Выделяем это слово. Найденное слово считается следующим токеном (сегментом). Указатель сдвигается вправо, сразу за конец этого слова.
4. Повторяем. Возвращаемся к шагу 2 для оставшейся части строки.
5. Завершаем. Алгоритм продолжается, пока вся строка не будет разбита на токены.

Ключевое допущение: Алгоритм жадный, он не пересматривает свои прошлые решения. Его эффективность сильно зависит от полноты и качества используемого словаря.

### Byte-Pair Encoding

Алгоритм BPE (Byte-Pair Encoding) — это метод сжатия данных, который также широко применяется для токенизации текста в задачах обработки естественного языка (NLP). Впервые он был предложен Филипом Гейджем в 1994 году как простой алгоритм сжатия, но позже адаптирован для построения словарей подслов (subword tokenization), например, в моделях GPT, BERT и других.

 
Основная идея: BPE позволяет разбивать текст на последовательности символов или байтов, а затем последовательно объединять наиболее часто встречающиеся пары в новые символы. Таким образом, алгоритм строит словарь, состоящий из отдельных символов и часто встречающихся их комбинаций (подслов). Это даёт возможность эффективно обрабатывать редкие и неизвестные слова, составляя их из известных подслов.

**Шаги алгоритма:**

1. Инициализация
* Разбить исходный текст на отдельные символы (или байты, если используется байтовая версия).
* Создать начальный словарь, содержащий все уникальные символы (например, буквы, знаки препинания, пробелы).
* Представить текст как последовательность этих символов.
2. Подсчёт частот пар
* Пройти по текущей последовательности и подсчитать, сколько раз каждая пара соседних символов (или байтов) встречается в тексте.
* Выбор наиболее частой пары
* Найти пару с максимальной частотой. Если несколько пар имеют одинаковую частоту, можно выбрать любую (обычно первую).
3. Объединение
* Создать новый символ, представляющий эту пару (например, если пара «a» и «b» встречается часто, вводим новый токен «ab»).
* Добавить новый символ в словарь.
* Заменить все вхождения этой пары в тексте на новый символ.
4. Повторение
* Повторять шаги 2–4 заданное количество раз (обычно до достижения желаемого размера словаря) или пока не перестанут появляться пары с частотой больше 1.

**Пример:**
Рассмотрим строку: "aaabdaaabac"
1. Начальное разбиение: a a a b d a a a b a c [a,b,c,d] 
2. Частоты пар: aa встречается 4 раза (позиции 1-2, 2-3, 6-7, 7-8), ab — 2 раза (3-4 и 8-9), bd — 1, da — 1, ba — 1, ac — 1. Самая частая — aa.
3. Объединяем aa в новый символ, обозначим его A [a,b,c,d,A]. Текст становится: A a b d A a b a c (заменяем все aa).
4. Снова считаем пары: Aa встречается 2 раза, Ab — 2 раза (после первых двух A и после A A перед b), bd — 1, dA — 1, Ab (снова), ba — 1, ac — 1. Теперь самая частая — Aa или Ab (обе по 2). Выберем Aa.
5. Объединяем Aa в новый символ B. Текст: B b d B b a c. [a,b,c,d,A (aa -> A),B (Aa )|('aa a') -> B)]
6. И так далее.

После нескольких итераций словарь будет содержать как отдельные символы, 
так и часто встречающиеся последовательности, что позволяет компактно представлять текст.


Посмотрим, как полный алгоритм работает на этом маленьком синтетическом корпусе, где мы явно отметили пробелы между словами: set_new_new_renew_reset_renew

Сначала мы разобьем корпус на слова, сохраняя начальные пробелы, вместе с их частотами; никакие слияния не будут выходить за пределы этих границ слов.
В результате получается следующий список из 4 слов и начальный словарь из 7 символов:

**корпус**

2 _ n e w

2 _ r e n e w

1 s e t

1 _ r e s e t

**словарь**

_,e,n,r,s,t,w

Алгоритм обучения BPE сначала подсчитывает все пары соседних символов: 
наиболее частой является пара **ne**, поскольку она встречается в **new** (частота 2) и **renew** (частота 2) — всего 4 вхождения. 
Затем мы объединяем эти символы, считая **ne** одним символом, и подсчитываем снова.

**корпус**

2 _ ne w

2 _ r e ne w

1 s e t

1 _ r e s e t

**словарь**

_,e,n,r,s,t,w,ne

Теперь наиболее частая пара — ne w (всего 4), которую мы объединяем.

**корпус**

2 _ new

2 _ r e new

1 s e t

1 _ r e s e t

**словарь**

_,e,n,r,s,t,w,ne,new

И т.д.

### BPE-кодировщик

После того как мы изучили наш словарь, для токенизации тестового предложения используется BPE-кодировщик. Кодировщик просто применяет к тестовым данным те объединения (мержи), которые мы извлекли из обучающих данных. Он применяет их «жадно», в том порядке, в котором мы их выучили. (Таким образом, частоты в тестовых данных не играют роли, учитываются только частоты из обучающих данных). Итак, сначала мы разбиваем каждое слово тестового предложения на символы. Затем мы применяем первое правило: замените каждое вхождение `n e` в тестовом корпусе на `ne`, затем второе правило: замените каждое вхождение `ne w` в тестовом корпусе на `new` и так далее.

К концу процесса многие объединения, по сути, просто воссоздают слова из обучающего набора. Но эти объединения также создают знание о морфемах, таких как префикс `re-` (который может появиться в, возможно, невиданных ранее комбинациях, например, *revisit* или *rearrange*), или морфема `new` без начального пробела (то есть внутри слова), которая может появиться в начале предложений или в словах, не встречавшихся при обучении, например, *anew*.

Конечно, в реальных условиях BPE выполняется с десятками тысяч объединений на очень большом входном корпусе, чтобы получить размер словаря в 50 000, 100 000 или даже 200 000 токенов. Результат таков, что большинство слов могут быть представлены как один токен, и только более редкие слова (а также неизвестные слова) должны будут представляться несколькими токенами. По крайней мере, для английского языка. Для многоязычных систем токены могут быть преимущественно английскими, оставляя меньше токенов для других языков, как мы обсудим ниже.

#### Применение в NLP
В современном NLP BPE используется для токенизации текста на подслова. 
Это решает проблему обработки неизвестных слов (out-of-vocabulary): 
любое слово можно разбить на известные подслова или даже отдельные символы. 
Размер словаря регулируется количеством операций слияния: 
чем больше слияний, тем крупнее токены (ближе к целым словам), и наоборот.

# Minimal Edit Distance (Расстояние Левенштейна)

## 1. История возникновения

Алгоритм вычисления минимального редакционного расстояния был предложен советским математиком **Владимиром Иосифовичем Левенштейном** в 1965 году. Первоначально он изучал вопросы теории кодирования и рассматривал операции вставки, удаления и замены символов как способы преобразования одной последовательности в другую. Эта метрика позже получила название **расстояние Левенштейна**.

Независимо схожие идеи разрабатывались в других областях:
- В биоинформатике для сравнения генетических последовательностей (алгоритмы Нидлмана–Вунша и Смита–Ватермана).
- В обработке естественного языка и поисковых системах для оценки схожести слов (исправление опечаток, автодополнение).

Сегодня алгоритм Minimal Edit Distance является фундаментальным инструментом во многих прикладных задачах, где требуется измерить «близость» двух строк.

## 2. Общая идея

**Минимальное редакционное расстояние** между двумя строками — это наименьшее количество элементарных операций, необходимых для превращения одной строки в другую. Обычно рассматриваются три операции:
- **Вставка** символа (insertion) (uv -> uav)
- **Удаление** символа (deletion) (uav -> uv)
- **Замена** одного символа на другой (substitution) (uav -> ubv)

Каждая операция имеет стоимость (обычно 1). Замена может иногда считаться как комбинация удаления и вставки, но в классическом расстоянии Левенштейна она стоит одну единицу.

Пример:  
Преобразуем строку `kitten` в `sitting`:
1. Замена `k` → `s` (1)
2. Замена `e` → `i` (1)
3. Вставка `g` в конец (1)
Итого расстояние = 3.

Алгоритм используется:
- В spell‑checker’ах для предложения исправлений.
- В системах распознавания речи для сравнения гипотез.
- В биоинформатике для выравнивания ДНК/РНК/белков.
- При анализе текстов для оценки степени плагиата или вариативности.

## 3. Описание алгоритма (динамическое программирование)

Классический способ вычисления — метод динамического программирования. Идея: построить матрицу `D` размером `(n+1) x (m+1)`, где `n = len(s1)`, `m = len(s2)`. Элемент `D[i][j]` хранит минимальное расстояние между префиксами `s1[:i]` и `s2[:j]`.

### Шаг 1. Инициализация

- `D[0][0] = 0` — пустые строки совпадают.
- Для `i = 1..n`: `D[i][0] = i` — чтобы получить пустую строку из префикса длины `i`, нужно удалить все символы.
- Для `j = 1..m`: `D[0][j] = j` — чтобы получить префикс длины `j` из пустой строки, нужно вставить все символы.

### Шаг 2. Заполнение матрицы

Для каждого `i` от 1 до `n` и каждого `j` от 1 до `m` вычисляем:

```python
# cost = 0, если s1[i-1] == s2[j-1], иначе 1 (стоимость замены)
# D[i][j] = min(
#     D[i-1][j] + 1,          // удаление символа s1[i-1]
#     D[i][j-1] + 1,          // вставка символа s2[j-1]
#     D[i-1][j-1] + cost      // замена (или совпадение)
# )
```


### Шаг 3. Результат

Искомое минимальное расстояние находится в ячейке `D[n][m]`.

### Восстановление последовательности операций

При желании можно восстановить сами операции, сохраняя, какой из трёх вариантов был выбран на каждом шаге, и затем пройти обратно от `D[n][m]` до `D[0][0]`.

## Пример вычисления

Рассмотрим строки: `s1 = "cat"`, `s2 = "cars"`.

Матрица `D` (размер 4×5):

|       |     | j=0 | j=1 | j=2 | j=3 | j=4 |
|-------|-----|-----|-----|-----|-----|-----|
|       |     |     | c   | a   | r   | s   |
| **i=0** |     | 0   | 1   | 2   | 3   | 4   |
| **i=1** | c   | 1   | 0   | 1   | 2   | 3   |
| **i=2** | a   | 2   | 1   | 0   | 1   | 2   |
| **i=3** | t   | 3   | 2   | 1   | 1   | 2   |

- `D[3][4] = 2` — минимальное расстояние равно 2.  
Одно из возможных преобразований: заменить `t` на `r` (или `t` на `r` и вставить `s` — но замена уже учтена).

## Сложность и модификации

- **Время работы**: O(n·m). Память также O(n·m), но можно сократить до O(min(n,m)) при хранении только двух строк матрицы.
- **Взвешенные операции**: иногда вводят разные стоимости для вставки, удаления и замены (например, замена может стоить 2, если считать её как два шага).
- **Расстояние Дамерау–Левенштейна**: добавляется операция транспозиции (перестановки двух соседних символов) стоимостью 1, что полезно для учёта опечаток вроде `ie` → `ei`.

```python
def levenshtein_distance(s1, s2):
    n, m = len(s1), len(s2)
    # Создаём матрицу (n+1) x (m+1)
    dp = [[0] * (m + 1) for _ in range(n + 1)]

    # Инициализация
    for i in range(n + 1):
        dp[i][0] = i
    for j in range(m + 1):
        dp[0][j] = j

    # Заполнение
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                cost = 0
            else:
                cost = 1
            
            dp[i][j] = min(
                dp[i - 1][j] + 1,        # удаление
                dp[i][j - 1] + 1,        # вставка
                dp[i - 1][j - 1] + cost  # замена
            )
    return dp[n][m]

# Пример использования
print(levenshtein_distance("kitten", "sitting"))  # 3
```
