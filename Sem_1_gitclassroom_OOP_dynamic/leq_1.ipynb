#### Задача (решается с преподавателем)

Написать функции __fibonacci(n: int) -> int__ вычисляющую числа Фибоначи по заданному $n$. Числа Фибоначи — это последовательность чисел, где каждое следующее число является суммой двух предыдущих чисел, задаётся формулой: 
$$ F_n = \begin{cases}
0, & n = 0 \\
1, & n = 1 \\
F_{n-1} + F_{n-2}, & n > 1
\end{cases} $$
Выведите 10 первых чисел Фибоначи списком.

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

if __name__ == "__main__":
    print([fibonacci(i) for i in range(10)])
```

# Объектно Ориентированное Программирование 

Вы уже знаете основные принципы ООП. Давайте посмотрим, как они реализуются в Python. Главное — понять синтаксис и философию языка, который часто следует принципу "явное лучше неявного".

1. Класс и \_\_init\_\_ (Конструктор)
Класс создаётся с помощью ключевого слова __class__. Для инициализации объекта используется магический метод \_\_init\_\_. Первый аргумент любого метода объекта — всегда self (ссылка на сам объект). Это явное указание, к чему мы обращаемся.

```python
class Car:
    # Конструктор
    def __init__(self, brand, model):
        # Атрибуты объекта (свойства)
        self.brand = brand
        self.model = model
        self._mileage = 0  # Начальное состояние. Одно подчёркивание _ — "защищённый" атрибут (соглашение).

    # Метод объекта
    def drive(self, distance):
        self._mileage += distance
        print(f"Машина проехала {distance} км. Общий пробег: {self._mileage} км")

# Создание объектов (экземпляров класса)
my_car = Car("Toyota", "Camry")
friend_car = Car("Tesla", "Model 3")

my_car.drive(150)  # Машина проехала 150 км. Общий пробег: 150 км
my_car._mileage
```

2. Инкапсуляция: "Защищённые" и "Приватные" атрибуты
Python полагается на соглашения, а не на строгие запреты.

- public_attribute — публичный атрибут, доступен отовсюду.

- _protected_attribute — "защищённый". Одно подчёркивание — соглашение для программиста: "не используй снаружи без крайней необходимости". Интерпретатор не ограничивает доступ.

- __private_attribute — "приватный". Два подчёркивания — включает name mangling (искажение имени). Интерпретатор меняет имя внутри класса на _ClassName__attribute, чтобы затруднить случайный доступ извне, но не делает его полностью недоступным.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner           # Публичный
        self._balance = balance      # Защищённый (внутренняя реализация)
        self.__pin = "1234"          # Приватный

    def get_balance(self):           # Публичный метод для доступа к защищённым данным
        return self._balance

acc = BankAccount("Иван", 1000)
print(acc.owner)        # OK
print(acc._balance)     # Доступ есть, но так делать не принято (нарушаем соглашение)
#print(acc.__pin)      # Ошибка! AttributeError
print(acc._BankAccount__pin) # Но так можно! (1234) — это и есть name mangling.
```

3. Наследование и Полиморфизм

- Наследование: классы наследуют свойства и методы от родительского класса.
- Полиморфизм: один класс может быть использован как класс наследуемого, так и класса родительского.

Синтаксис прост: в скобках указываем родительский класс. Методы переопределяются автоматически. Можно обращаться к родительскому методу через super().

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print("...")

class Dog(Animal):  # Наследование
    def make_sound(self):  # Переопределение метода (Полиморфизм)
        print("Гав!")

    def fetch(self):       # Собственный метод подкласса
        print(f"{self.name} принёс палку!")

class Cat(Animal):
    def make_sound(self):
        print("Мяу!")

animals = [Dog("Барсик"), Cat("Мурка")]  # Разные типы в одном списке

for animal in animals:
    animal.make_sound()  # Один интерфейс — разное поведение
# Вывод: Гав! Мяу!
```

```python
class Animal:
    def __init__(self, name, species):
        print("Инициализация Animal")
        self.name = name
        self.species = species
        self.is_alive = True
    
    def info(self):
        return f"{self.name} ({self.species})"

class Cat(Animal):
    def __init__(self, name, breed, age):
        # Вызываем конструктор родительского класса Animal
        super().__init__(name, species="Кошка")
        print("Инициализация Cat")
        
        # Дополнительные атрибуты только для Cat
        self.breed = breed
        self.age = age
        
    # Переопределяем метод info, но используем родительскую реализацию
    def info(self):
        # Получаем базовую информацию от родителя
        base_info = super().info()
        return f"{base_info}, порода: {self.breed}, возраст: {self.age} лет"

# Создаем объект
my_cat = Cat("Мурзик", "Британская", 3)
print(my_cat.info())
```

4. Особенности Python: Магические методы (Dunder methods)

Это методы с двойным подчёркиванием вначале и в конце (\_\_len\_\_, \_\_str\_\_ и т.д.). Они позволяют объектам интегрироваться в язык. Например, print(obj) автоматически вызывает obj.\_\_str\_\_().

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):  # Для красивого строкового представления (для человека)
        return f"'{self.title}' от {self.author}"

    def __len__(self):  # Позволяет использовать функцию len() для объекта
        return 300  # Условно, длина книги в страницах

book = Book("1984", "Дж. Оруэлл")
print(book)    # Вызовется __str__: '1984' от Дж. Оруэлл
print(len(book)) # Вызовется __len__: 300
```

Перегрузка операторов — это возможность задать поведение объектов вашего класса при использовании стандартных операторов Python (+, -, *, /, ==, < и т.д.). Это делает код интуитивно понятным и похожим на встроенные типы.

Когда вы пишете a + b, Python делает следующее:

- Смотрит на тип объекта a.
- Ищет у него магический метод \_\_add\_\_.
- Вызывает a.\_\_add\_\_(b).
- Если у a нет \_\_add\_\_ или он возвращает специальное значение NotImplemented, Python пробует b.\_\_radd\_\_(a) (правосторонний вариант).

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        elif isinstance(other, (int, float)):
            return Vector(self.x + other, self.y + other)
        else:
            return NotImplemented

    def __radd__(self, other):
        return self.__add__(other)

v1 = Vector(2, 4)
v2 = Vector(3, 1)

print(v1 + v2)          
print(v1 + 5)           
print(5 + v1)           
```

Аналогично можно переопределять функции \_\_mul\_\_ и \_\_rmul\_\_. 

Немного поговорим про итераторы. В Python всё является объектом, включая итераторы и генераторы. 
Протокол итерации — это классический пример полиморфизма в ООП: любой объект, реализующий определенные методы, может использоваться в цикле for.

```python
class Counter:
    def __init__(self, start, end):
        self.current = start
        self.end = end
    
    def __iter__(self): # это сделает объект итерируемым
        return self
    
    def __next__(self): # это позволит искать следующий элемент
        if self.current >= self.end:
            raise StopIteration
        result = self.current
        self.current += 1
        return result

# Использование
counter = Counter(1, 4)
for num in counter:  # Автоматически вызывает __iter__, затем __next__
    print(num)  

# Можно вручную
counter = Counter(1, 4)
iterator = iter(counter)  # Вызывает __iter__
print(next(iterator))  
print(next(iterator))  
print(next(iterator))  
# print(next(iterator))  # StopIteration
```

# Что такое yield?

yield — это ключевое слово в Python, которое превращает обычную функцию в генератор. Генератор — это специальный итератор, который генерирует значения по одному по мере необходимости, а не все сразу.

```python
def get_numbers_list(n):
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result  

numbers = get_numbers_list(100)  # Занимает много памяти!
print(numbers)
```

```python
def get_numbers_generator(n):
    for i in range(n):
        yield i**2  

numbers = get_numbers_generator(100)  # Почти не занимает памяти!
print(numbers)
print(next(numbers))
```

```python
print(next(numbers))
```

```python
for i in get_numbers_generator(100):
    print(i)
```

Когда функция содержит yield:

1. При вызове она не выполняется сразу, а возвращает объект-генератор
2. Код функции выполняется только при вызове next() или в цикле for
3. При достижении yield функция приостанавливается и возвращает значение
4. При следующем вызове выполнение продолжается с того же места

```python
def simple_generator():
    print("Начало")
    yield 1
    print("Продолжение")
    yield 2
    print("Конец")
    yield 3

gen = simple_generator()  # Ничего не выводится!

print(next(gen))  # "Начало", затем 1
print(next(gen))  # "Продолжение", затем 2
print(next(gen))  # "Конец", затем 3
#print(next(gen))  # StopIteration - генератор закончился
```

Ключевые преимущества генераторов
1. Экономия памяти — не хранят все элементы сразу
2. Ленивые вычисления — вычисляют значения по мере необходимости
3. Бесконечные последовательности — можно генерировать бесконечно
4. Сохранение состояния — функция "помнит" где остановилась

Когда использовать генераторы?

Используйте генераторы, когда:

1. Работаете с большими объемами данных
2. Нужна ленивая загрузка
3. Работаете с бесконечными последовательностями

Не используйте генераторы, когда:
1. Нужен произвольный доступ к элементам (по индексу)
2. Нужно несколько раз пройти по одним данным

# Динамическое програмирование

Динамическое программирование — это способ решения сложных задач путём разбиения их на более простые подзадачи, результаты которых мы запоминаем и переиспользуем. Если вы когда-то решали задачу "найти 100-е число Фибоначчи" и сталкивались с тем, что наивная рекурсия работает очень медленно — вы уже встречались с проблемой, которую решает ДП.

#### Основная идея экономить вычисления

Например, задача про числа Фибоначи. 

__Вопрос:__ В чём проблема решения задачи про числа Фибоначи? 

__Ответ:__ Мы много раз вычисляем одно и то-же. Например, как устроенно вычисление $F_5$: 

$$ F_5 = F_{4} + F_{3} = F_{3} + F_{2} + F_{3} =
(\text{тут мы начали 3-ий расписывать}) \\ F_{2} + F_{1} + F_{2} + F_3 = F_{1} + F_{0} + F_{1} + F_{2} + F_3 = \\
= 1 + F_{0} + F_{1} + F_{2} + F_3 = 1 + 0 + F_{1} + F_{2} + F_3 = 1 + 0 + 1 + F_{2} + F_3 (\text{а тут мы закончили}) = \\
= 1 + 0 + 1 + F_{1} + F_0 + F_3 = 1 + 0 + 1 + 1 + F_0 + F_3 = 1 + 0 + 1 + 1 + 0 + F_3 = \\ (\text{тут опять начали 3-ий расписывать}) 
= 1 + 0 + 1 + 1 + 0 + F_2 + F_1 = 1 + 0 + 1 + 1 + 0 + F_1 + F_0 + F_1 = \\
= 1 + 0 + 1 + 1 + 0 + 1 + F_0 + F_1 = 1 + 0 + 1 + 1 + 0 + 1 + 0 + F_1 = \\
= 1 + 0 + 1 + 1 + 0 + 1 + 0 + 1 (\text{и тут наконец закончили}) = 5 $$

Итого делаем кучу ненужныхных вычислений. 
Как это поменять? 

```python
def fibonacci(n):
    fib_set = []
    fib_set.append(0)
    fib_set.append(1)
    for i in range(2,n+1):
        fib_set.append(fib_set[i-1] + fib_set[i-2])
    return fib_set[n]

print([fibonacci(i) for i in range(10)])
```

Как теперь устроено вычисление $F_5$: 
$$ F_0 = 0 \quad F_1 = 1 \quad F_2 = 1 \quad F_3 = 2 \quad F_4 = 3 \quad F_5 = 5 $$
Выиграли? Как минимум не считает $F_3$ несколько раз. 

__Вопрос:__ Можно ли это объединить с ленивыми вычислениями?
__Ответ:__ ДА!

```python
def fibonacci_lazy():
    yield 0
    yield 1
    fib_set = [0,1]
    while True:
        fib_set.append(fib_set[-1] + fib_set[-2])
        yield fib_set[-1]

a = fibonacci_lazy()
print(a)
```

#### Задача

Имеется поле $n \times m$ (n - количество строк, m - количество столбцов), сколькими способами можно добраться королём из нижней левой клетки до правой верхней клетки.

```python
def king_moves(n,m):
    king_set = [[0] * m for i in range(n)]
    for i in range(n):
        for j in range(m):
            if i == 0 and j == 0:
                king_set[0][0] = 1
            elif i == 0:
                king_set[0][j] = 1  
            elif j == 0:
                king_set[i][0] = 1
            else:
                king_set[i][j] = king_set[i-1][j] + king_set[i][j-1]
            print(king_set)
    return king_set[n-1][m-1]

def king_moves_2(n,m):
    king_set = []
    for i in range(n):
        for j in range(m):
            if i == 0 and j == 0:
                king_set.append(1)
            elif i == 0:
                king_set.append(1)
            elif j == 0:
                king_set.append(1)
            else:
                king_set.append(king_set[-1] + king_set[-n])
            print(king_set)
    return king_set[n*m-1]

print(king_moves(3,3))
print(king_moves_2(3,3))
```

Но иногда вы знаете математику ...

```python
import math 

def king_moves_3(n,m):
    return math.factorial(n+m-2) / (math.factorial(n-1) * math.factorial(m-1))

print(int(king_moves_3(3,3)))
```
